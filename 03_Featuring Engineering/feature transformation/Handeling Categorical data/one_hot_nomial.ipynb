```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop='first',dtype=np.int32)
ohe.fit(xtrain[['fuel','owner']])
xtrain_new=ohe.transform(xtrain[['fuel','owner']]).toarray()
xtest_new=ohe.transform(xtest[['fuel','owner']]).toarray()
xtrain_new.shape
```

**4.One Hot Encoding with Top categories**

```python
count=df['brand'].value_counts()
count
```

```python
df['brand'].nunique()
```

```python
threshold=100
```

```python
replace=count[count<=threshold].index
###gives all brands having cars less than equal to threshold
```

```python
dff=pd.get_dummies(df['brand'].replace(replace,'others'),dtype=int)
```
