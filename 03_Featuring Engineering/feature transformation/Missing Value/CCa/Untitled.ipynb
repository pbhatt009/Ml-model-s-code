```python
value=df.isnull().mean()*100
```

```python
### we will remove thosse rows which have mising values <5%
```

```python
cols = [col for col in df.columns if df[col].isnull().mean() > 0.0 and df[col].isnull().mean() < 0.05]
cols
```

```python
newdf=df[cols].dropna()
newdf
```

## Checking ratio for catgorical data

```python
def ratio(colname):
    before=[]
    after=[]
    for i in df[f'{colname}'].value_counts():
        before.append(i/len(df))
    for i in newdf[f'{colname}'].value_counts():
        after.append(i/len(newdf))
    print(pd.DataFrame(np.column_stack((before,after)),columns=[' before','after'],index=[df[f'{colname}'].value_counts().index]))
    
 
    
```

```python
ratio('enrolled_university')
```
