```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('titanic_toy.csv')
```

```python
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(df.drop('Survived',axis=1),df['Survived'],test_size=0.2,random_state=42)
```

```python
from sklearn.impute import KNNImputer
```

```python
knn=KNNImputer( n_neighbors=5, weights='distance')
```

```python
knn.fit(xtrain)
xtrain_imp=knn.transform(xtrain)
xtest_imp=knn.transform(xtest)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
dlf=LogisticRegression()
dlf.fit(xtrain_imp,ytrain)
```

```python
predct=dlf.predict(xtest_imp)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,predct)
```
