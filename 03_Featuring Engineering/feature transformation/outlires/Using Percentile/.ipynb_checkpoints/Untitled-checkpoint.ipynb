## we will do outlire finding in height
df['Height'].describe()

```python
ul=df['Height'].quantile(0.99)
ll=df['Height'].quantile(0.01)
```

```python
outlires=df[(df['Height']<=ll)|(df['Height']>=ul)]
outlires.shape
```

##  Capping(vinserization) ##

```python
height_new=pd.DataFrame(np.where((df['Height']>=ll)&(df['Height']<=ul),df['Height'],np.where(df['Height']>ul,ul,ll)))
```

```python
sns.boxplot(height_new)
```
