```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('customer.csv')
df
#review and education -ordinal encoding
#purchased-label encoding 
#gender-one hot
```

```python
df['review'].describe()
```

```python
df=df.iloc[:,2:]
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.2,random_state=1)
```

```python
ytrain=pd.DataFrame(ytrain,columns=['purchased'])
ytest=pd.DataFrame(ytest,columns=['purchased'])
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(xtrain)
xtrain=pd.DataFrame(oe.transform(xtrain),columns=xtarin.columns)
xtest=pd.DataFrame(oe.transform(xtest),columns=xtest.columns)

```

**Label Encoding for output**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
le.fit(ytrain['purchased']) # we can not decide order
le.classes_
        
```

```python
ytrain=le.transform(ytrain['purchased'])
ytest=le.transform(ytest['purchased'])    
```
