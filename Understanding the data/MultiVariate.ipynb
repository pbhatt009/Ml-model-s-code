**2.BarPlot(NUmerical-Catogarical)**

**3.Box Plot(Numerical-cat)**

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['2urvived'])
#0-female
```

**4.displot(nu-cat)**

```python
# print(titanic)
# print(titanic[titanic['2urvived']==1.0]['Fare'])

sns.histplot(titanic[titanic['2urvived'] == 1.0]['Age'], kde=True,stat='probability')
# gives the sreies of age of survived people
sns.histplot(titanic[titanic['2urvived'] == 0.0]['Age'], kde=True,stat='probability')
```

```python
sns.kdeplot(titanic[titanic['2urvived'] == 1.0]['Age'],color='Green')
sns.kdeplot(titanic[titanic['2urvived'] == 0.0]['Age'])
```

**5.Heat Map (Cat-cat)**

```python
pd.crosstab(titanic['Pclass'],titanic['2urvived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['2urvived']))
```

```python
pclass=(titanic.groupby('Pclass'))
```

```python
pclass.size()
```

```python
pclass.mean()
```

```python
(pclass.mean()['2urvived']*100).plot(kind="bar",ylabel="%survived")
```

```python
(titanic.groupby('Sex').mean()['2urvived']*100).plot(kind="bar",ylabel="%survived")
```
